### define function to compute percentage

```r
prop <- function(col_name, pop_total) {
    the_list <- list()
    for (i in col_name) {
        x <-  (i * 100)/pop_total
        x <- format(round(x, 2), nsmall = 2)
        y <- paste (as.character(i), ' (', as.character(x), ')', sep = '')
        the_list <- append(the_list, y)
    }
    the_list
        }
```

## DEFINE FUNCTION FOR RELATIVE PROPORTION

```r
### function to compute relative proportion and bootstrapped confidence interval
rel.prop <- function (data) {
    i <- data[data$group == "sample_p1" & data$allele == 1, ] %>% nrow()
    b <- data[data$group == "sample_p1", ] %>% nrow()
    q <- data[data$group == "sample_p2" & data$allele == 1, ] %>% nrow()
    the_rest <- data[data$group == "sample_p2", ] %>% nrow()

    # prevalence_ratio between sub-population of interest and the others
    p1 <- i/b
    p2 <- q/the_rest

    ### bootstrap confidence interval
    rel_prop_bootstrap <- c()
    n <- nrow(data)

    for(i in c(1:1000)) {
        set.seed(42*i)  # This controls the randomization process
        boot <- sample(n, n, replace=TRUE)
        resampled_df <- data[boot, ]  # Selects SRSWR
        resampled_i <- resampled_df[resampled_df$group == "sample_p1" & resampled_df$allele == 1, ] %>% nrow()
        resampled_b <- resampled_df[resampled_df$group == "sample_p1", ] %>% nrow()
        resampled_q <- resampled_df[resampled_df$group == "sample_p2" & resampled_df$allele == 1, ] %>% nrow()
        resampled_the_rest <- resampled_df[resampled_df$group == "sample_p2", ] %>% nrow()
        
        # resampled prevalence_ratio between sub-population of interest and the others
        ## report bootstrapped prevalence ratios as missing value if denominator is 0 (to avoid infinity values)
        resampled_p1 <- resampled_i/resampled_b
        resampled_p2 <- resampled_q/resampled_the_rest
        if (resampled_p2 == 0) {
            rel_prop_bootstrap[i] <- NA
        } else if (resampled_p2 > 0) {
            rel_prop_bootstrap[i] <- resampled_p1/resampled_p2
        }
    }


    ## report prevalence ratio as missing value if denominator is 0 (to avoid infinity values)
    prevalence_ratio <- p1/p2
    if (p2 == 0) {
        prevalence_ratio <- NA
    } else if (p2 > 0) {
        prevalence_ratio <- p1/p2
    }
    lowerCI <- quantile(rel_prop_bootstrap, 0.025, na.rm = T)
    upperCI <- quantile(rel_prop_bootstrap, 0.975, na.rm = T)

    
    # round off to 2dp
    prevalence_ratio <- format(round(prevalence_ratio, 2), nsmall = 2) 
    lowerCI <- format(round(lowerCI, 2), nsmall = 2) 
    upperCI <- format(round(upperCI, 2), nsmall = 2) 
        
        
    concat_the_string <- str_c (prevalence_ratio, ' (', lowerCI, ', ', upperCI, ')')
    return(concat_the_string)
}
```

## DEFINE FUNCTION FOR POPULATION VS REST OF THE WORLD

```r
### function to compute confidence interval 
conf_int_v_ALL <- function (b, column_name) {
    # b = total number of alleles in specific sub-population; 
    # a = total number of alleles in global population; 
    to_list (for (item in Map(list, column_name, cross_continent$'ALL (n = 5008)')) {
        i = item[[1]] # i = frequency of queried allele in specific sub-population
        j = item[[2]] # frequency of queried allele in global population

        a = 5008
        the_rest <- a - b # total number of alleles in sub-populations other than the one being queried
        q = j - i # frequency of queried allele among `the_rest`


        sample_p1 <- c(rep(1, i), rep(0, b-i))
        sample_p2 <- c(rep(1, q), rep(0, the_rest-q))

        sample_p1 <- sample(sample_p1, replace = FALSE)
        sample_p2 <- sample(sample_p2, replace = FALSE)

        sample_p1 <- data.frame(allele = sample_p1, group = "sample_p1")
        sample_p2 <- data.frame(allele = sample_p2, group = "sample_p2")
        both_df <- rbind(sample_p1, sample_p2)
        
        
        rel.prop(both_df)

    }
            )
}
```
